```python
# Importamos las librerias a usar
import pandas as pd
import json
import ast
```

```python
# Creo un lista donde se depositaran los datos
itemsList = []
# Abro un archivo, señalizando que tiene un encoding 'utf-8'
with open("australian_users_items.json", "r", encoding="utf-8") as f:
  # Leo cada linea con el bucle for
  for line in f.readlines():
      # lee todas las líneas del archivo y las coloca en una lista de cadenas.
    itemsList.append(ast.literal_eval(line))
# Convierto el archivo a un data frame
dfItems = pd.DataFrame(itemsList)
# Abro los archivos
```

```python
#Eliminar las columnas menos relevantes del Dataframe Items
columnasAEliminar = ['user_url']
dfItems = dfItems.drop(columns=columnasAEliminar)
dfItems
#Compruebo cantiadad de nulos
dfItems.isna().sum()
# Explorar la columna items
dfItems = dfItems.explode('items')

#Compruebo cantidad de nulos
dfItems.isna().sum()
#Elimino nulos
dfItems= dfItems.dropna(subset=['items'])
```

```python
#Lectura en Google collab dado la falta del RAM

for elemento,diccionario in dfItems['items'].head().items():
    df=pd.DataFrame(columns =  diccionario.keys())
new_rows = []

# Iterate over the dictionaries in the 'reviews' column
for elemento, diccionario in dfItems['items'].items():
    new_row = pd.DataFrame(diccionario, index=[elemento])  # Create a new row DataFrame
    new_rows.append(new_row)  # Append the new row to the list

# Concatenate the list of new rows with the original DataFrame
nuevoDf = pd.concat([df, *new_rows])
#Juntar la nueva columnas con la anterior 
dfItems = dfItems.join(nuevoDf)

#Eliminar columnas
dfItems=dfItems.drop_duplicates(subset=['user_id','item_name'])
```
